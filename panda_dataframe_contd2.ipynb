```python
#dropna : This method can be applied on a series and dataframe as well. 

df1[0].dropna()
```

```python
#dropna : If this method is applied on a dataframe, then it will delete all the rows, which have missing values even in a single
#column by default
df1.dropna()
```

```python
#however, we can use how parameter to customize this functionality. 
df1.dropna(how='all')
```

```python
df1
```

```python
#fillna : fillna method is used to fill the missing values with another value. It can be filled up with mean, median, mode, 
#backfill, forward fill etc. 
#filling the missing values with the mean of the columns.
df1.fillna(df1.mean())
```

```python
#filling the missing values with the mean of the rows.
df1.fillna(df1.mean(axis=1))
```

```python
#fillna is typically used on a series only to be more meaningful. 
df1[0].fillna(df1[0].mean())
```

```python
df1
```

```python
#ffill method is used to fill the nan values with the values of the column above it. However, if nan values exists in the 
#first row itself, nan values will remain as it is. 
df1.fillna(method='ffill')
```

```python
#bfill method is used to fill the nan values with the back values of the column below it. However, if nan values exists in the 
#last row itself, nan values will remain as it is. 
df1.fillna(method='bfill')
```

```python
#Initializing the dataframe
df = pd.DataFrame([[10,30],
                  [12,29],
                  [31,50]],
                 columns=['min_temp','max_temp'])
#printing the dataframe
print(df)
#defining a method 
def calc_diff(row):
    
    return row['max_temp']-row['min_temp']

#applying the method on the row of the dataframe
print(df.apply(calc_diff,axis=1))
```
